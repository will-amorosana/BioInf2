```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Load in the data and clean the percentages
gs1 = pd.read_csv('data/panther_res_1.txt', names=['index', 'name', 'count', 'weird', 'percent'], sep='\t')
gs2 = pd.read_csv('data/panther_res_2.txt', names=['index', 'name', 'count', 'weird', 'percent'], sep='\t')
gs3 = pd.read_csv('data/panther_res_3.txt', names=['index', 'name', 'count', 'weird', 'percent'], sep='\t')
for i in [gs1,gs2,gs3]:
    i['percent'] = i['percent'].str.rstrip('%').astype('float')
```

```python
labels = ['Translation Regulator', 'Transcription Regulator', 'Molecular Transducer',
             'Binding', 'Structural Molecule', 'Molecular Function', 'Catalytic', 'Transporter']
gs1_counts = gs1['percent'] 
gs2_counts = gs2['percent']
gs3_counts = gs3['percent']

fig = plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')

x = np.arange(len(labels))  # the label locations
width = 0.3  # the width of the bars

rects1 = plt.bar(x - width, gs1_counts, width, label='Group 1')
rects2 = plt.bar(x, gs2_counts, width, label='Group 2')
rects3 = plt.bar(x + width, gs3_counts, width, label='Group 3')
# Add some text for labels, title and custom x-axis tick labels, etc.
plt.ylabel('% of genes')
plt.title('Gene Distribution per Gene Score Group')
plt.xticks(x, labels, rotation=70)
plt.legend()
plt.margins(0.3)
fig.subplots_adjust(bottom=0.3)

fig.show()
fig.savefig('imgs/2_5_ext.pdf')
```
